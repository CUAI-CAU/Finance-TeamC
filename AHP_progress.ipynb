```python
def get_ahp_weight(csv_name):
    df = pd.read_csv(csv_name)
    positive = list(df['sentimental_score_positive'])[:-1]
    label = list(df['topic_label'])[:-1]
    a,b,c,d,e,f = positive[0], positive[1], positive[2], positive[3], positive[4], positive[5]
    end_list = [a,b,c,d,e,f] 
    print(len(positive),'개의 토픽 존재')
    if len(positive)>6:
        g = positive[6]
        end_list.append(g)
    if len(positive)>7:
        h = positive[7]
        end_list.append(h)
    if len(positive)>8:
        i = positive[8]
        end_list.append(i)
    if len(positive)>9:
        j = positive[9]
        end_list.append(j)
    if len(positive)>10:
        k = positive[10]
        end_list.append(k)
    if len(positive)>11:
        l = positive[11]
        end_list.append(l)
    if len(positive)>12:
        m = positive[12]
        end_list.append(m)
    if len(positive)>13:
        n = positive[13]
        end_list.append(n)
    if len(positive)>14:
        o = positive[14]
        end_list.append(o)
    #print(end_list)
    final_comp_matrix = comp_matrix_compute(end_list)
    col_sums =  final_comp_matrix.sum(axis=0) # 열의 합 print(‘열의 합 : {}'.format(col_sums)) 
    np02 =  final_comp_matrix/col_sums # 정규화(동일한 열에 있는 col_sumns의 값으로 나눠줍니다.) 
    #print('정규화 : {}'.format(np02))
    row_avg = np02.mean(axis=1) # 속성별 중요도, 가중치(mean(axis=1)은 행의 평균을 계산합니다.) 
    print('가중치 : {}'.format(row_avg))
    
    topic_weight = pd.DataFrame({'topic label': label ,'weight':row_avg})
    new_filename = csv_name + 'ahp_result.csv'
    topic_weight.to_csv(new_filename, index=False)
    return topic_weight
```
